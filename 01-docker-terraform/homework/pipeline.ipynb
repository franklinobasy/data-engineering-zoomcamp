```python
import pandas as pd
from time import time
```

```python
from sqlalchemy import create_engine
```

```python
user, password, host, port, db = "root", "root", "localhost", "5432", "ny_taxi"
engine = create_engine(f'postgresql://{user}:{password}@{host}:{port}/{db}')
```

```python
tripdata = pd.read_csv("green_tripdata_2019-09.csv", iterator=True, chunksize=20000)
df = next(tripdata)
df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
table_name = "trips"
df.head(n=0).to_sql(name=table_name, con=engine, if_exists='replace')
df.to_sql(name=table_name, con=engine, if_exists='append')
```

```python
while True: 
    try:
        t_start = time()

        df = next(tripdata)

        df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
        df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)

        df.to_sql(name=table_name, con=engine, if_exists='append')

        t_end = time()

        print('inserted another chunk, took %.3f second' % (t_end - t_start))

    except StopIteration:
        print("Finished ingesting data into the postgres database")
        break
```

```python
taxi_zones = pd.read_csv("taxi+_zone_lookup.csv")
taxi_zones.head(n=0).to_sql(name="taxi_zones", con=engine, if_exists='replace', index=False)
taxi_zones.to_sql(name="taxi_zones", con=engine, if_exists='append', index=False)
```
